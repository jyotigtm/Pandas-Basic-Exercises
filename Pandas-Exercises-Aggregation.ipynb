### Step 3. Assign it to a variable called users.

```python

users = pd.read_table('https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user', 
                      sep='|', index_col='user_id')
```

```python
users.head()
```

### Step 4. Discover what is the mean age per occupation

```python

round(users.age.mean())
```

```python
round(users.groupby('occupation').age.mean())
```

### Step 5. Discover the Male ratio per occupation and sort it from the most to the least.

Use numpy.where() to encode gender column.

```python
F = users.occupation.where(users.gender == 'F').count()
```

```python
M = users.occupation.where(users.gender == 'M').count()
```

```python
M_ratio = F/(F+M)
M_ratio
```

```python
def gender_to_numeric(x):
    if x == 'M':
        return 1
    if x == 'F':
        return 0

users['gender_n'] = users['gender'].apply(gender_to_numeric)


a = users.groupby('occupation').gender_n.sum() / users.occupation.value_counts()

a.sort_values(ascending = False)
```

```python
new1 = users.pivot_table(index='occupation', columns='gender', aggfunc='size', fill_value=0)
sums = new1[['F', 'M']].sum(axis=1)
new1['Ratio_of_Male'] = new1['M']/sums
     
new1
```

### Step 6. For each occupation, calculate the minimum and maximum ages

```python
users.groupby('occupation').age.agg(['min', 'max'])
```

### Step 7. For each combination of occupation and gender, calculate the mean age

```python
users.groupby(['occupation', 'gender']).age.mean()
```

```python
round(users.groupby(['occupation']).apply(lambda x: x[(users.gender=='M')].age.mean()))
```

```python
round(users.groupby(['occupation']).apply(lambda x: x[(users.gender=='F')].age.mean()))
```

### Step 8.  For each occupation present the percentage of women and men

```python
new = users.pivot_table(index='occupation', columns='gender', aggfunc='size', fill_value=0)
sums = new[['F', 'M']].sum(axis=1)
new['%ofwomen'] = round(new['F']/sums * 100)
new['%ofmen'] = round(new['M']/sums * 100)

print(new)
```
